```python


# Step 3: Univariate Analysis

# Question 1: Overall distribution of purchases (Incidence)
plt.figure(figsize=(6, 4))
sns.countplot(x='Incidence', data=data)
plt.title('Incidence Distribution')
plt.xlabel('Purchase Incidence')
plt.ylabel('Count')
plt.show()
```

```python

# Question 2: Number of brands customers typically buy (Brand and Quantity)
brand_count = data['Brand'].value_counts()
quantity_count = data['Quantity'].value_counts()

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
sns.barplot(x=brand_count.index, y=brand_count.values)
plt.title('Brand Count')
plt.xlabel('Brand')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
sns.barplot(x=quantity_count.index, y=quantity_count.values)
plt.title('Quantity Count')
plt.xlabel('Quantity')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python

# Step 4: Bivariate Analysis

# Group the data by ID and calculate the mean of the grouped data
grouped_data = data.groupby('ID').mean()

# Question 3: Relationship between current and previous purchases (Brand and Last_Inc_Brand)
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Brand', y='Last_Inc_Brand', data=grouped_data)
plt.title('Current vs. Previous Brand Purchase')
plt.xlabel('Current Brand')
plt.ylabel('Previous Brand')
plt.show()
```

```python

# Question 4: Variation of prices among different brands (Price_1 to Price_5)
price_cols = ['Price_1', 'Price_2', 'Price_3', 'Price_4', 'Price_5']
plt.figure(figsize=(8, 5))
sns.boxplot(data=grouped_data[price_cols])
plt.title('Price Distribution for Different Brands')
plt.xlabel('Brand')
plt.ylabel('Price')
plt.show()
```
