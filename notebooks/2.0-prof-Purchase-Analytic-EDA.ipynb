# Purchase Data - Legend

The dataset consists of valuable information regarding the purchases of chocolate candy bars made by 500 individuals in a physical 'FMCG' store over a 2-year period. The data has been collected through loyalty cards used at checkout and has undergone preprocessing, resulting in no missing values. Additionally, to safeguard customer privacy, the dataset's volume has been restricted and anonymized.

## Variable Descriptions

1. **ID**: Numerical - Shows a unique identifier of a customer.
2. **Day**: Numerical - Represents the day when the customer visited the store.
3. **Incidence**: Categorical {0,1} - Indicates the purchase incidence.
    - 0: The customer has not purchased an item from the category of interest.
    - 1: The customer has purchased an item from the category of interest.
4. **Brand**: Categorical {0,1,2,3,4,5} - Shows the brand ID that the customer has purchased.
    - 0: No brand was purchased.
    - 1, 2, 3, 4, 5: Brand ID.
5. **Quantity**: Numerical (Integer) - Represents the number of items bought by the customer from the product category of interest.
6. **Last_Inc_Brand**: Categorical {0,1,2,3,4,5} - Indicates the brand ID that the customer purchased on their previous store visit.
    - 0: No brand was purchased.
    - 1, 2, 3, 4, 5: Brand ID.
7. **Last_Inc_Quantity**: Numerical (Integer) - Represents the number of items bought by the customer from the product category of interest during their previous store visit.
8. **Price_1**: Numerical (Real) - Represents the price of an item from Brand 1 on a particular day.
9. **Price_2**: Numerical (Real) - Represents the price of an item from Brand 2 on a particular day.
10. **Price_3**: Numerical (Real) - Represents the price of an item from Brand 3 on a particular day.
11. **Price_4**: Numerical (Real) - Represents the price of an item from Brand 4 on a particular day.
12. **Price_5**: Numerical (Real) - Represents the price of an item from Brand 5 on a particular day.
13. **Promotion_1**: Categorical {0,1} - Indicates whether Brand 1 was on promotion on a particular day.
    - 0: There is no promotion.
    - 1: There is a promotion.
14. **Promotion_2**: Categorical {0,1} - Indicates whether Brand 2 was on promotion on a particular day.
    - 0: There is no promotion.
    - 1: There is a promotion.
15. **Promotion_3**: Categorical {0,1} - Indicates whether Brand 3 was on promotion on a particular day.
    - 0: There is no promotion.
    - 1: There is a promotion.
16. **Promotion_4**: Categorical {0,1} - Indicates whether Brand 4 was on promotion on a particular day.
    - 0: There is no promotion.
    - 1: There is a promotion.
17. **Promotion_5**: Categorical {0,1} - Indicates whether Brand 5 was on promotion on a particular day.
    - 0: There is no promotion.
    - 1: There is a promotion.
18. **Sex**: Categorical {0,1} - Represents the biological sex (gender) of a customer.
    - 0: Male.
    - 1: Female.
19. **Marital Status**: Categorical {0,1} - Indicates the marital status of a customer.
    - 0: Single.
    - 1: Non-single (divorced / separated / married / widowed).
20. **Age**: Numerical (Integer) - Represents the age of the customer in years, calculated as the current year minus the year of birth of the customer at the time of dataset creation.
    - Min value: 18 (the lowest age observed in the dataset).
    - Max value: 75 (the highest age observed in the dataset).
21. **Education**: Categorical {0,1,2,3} - Represents the level of education of the customer.
    - 0: Other / Unknown.
    - 1: High school.
    - 2: University.
    - 3: Graduate school.
22. **Income**: Numerical (Real) - Represents the self-reported annual income in US dollars of the customer.
    - Min value: 38,247 (the lowest income observed in the dataset).
    - Max value: 309,364 (the highest income observed in the dataset).
23. **Occupation**: Categorical {0,1,2} - Represents the category of occupation of the customer.
    - 0: Unemployed / Unskilled.
    - 1: Skilled employee / Official.
    - 2: Management / Self-employed / Highly qualified employee / Officer.
24. **Settlement Size**: Categorical {0,1,2} - Represents the size of the city that the customer lives in.
    - 0: Small city.
    - 1: Mid-sized city.
    - 2: Big city.

The provided dataset and variable descriptions will be utilized to conduct Exploratory Data Analysis (EDA) and derive meaningful insights into customer buying behavior for chocolate candy bars. The analysis aims to facilitate data-driven decision-making for businesses operating in this domain.

# Exploratory Data Analysis (EDA) and Visualization

## Problem Statement

The dataset contains information about the purchases of chocolate candy bars made by 500 individuals over a period of 2 years in a physical 'FMCG' store. The data has been collected through loyalty cards used at checkout and has been preprocessed with no missing values.

The objective of this analysis is to gain valuable insights into the purchasing behavior of customers and identify patterns or trends that can inform business decision-making.

## Exploratory Data Analysis

### Question 1: Overall Distribution of Purchases (Incidence)

To start our analysis, we visualize the overall distribution of purchases based on the 'Incidence' variable, which indicates whether a customer has purchased an item from the category of interest (1) or not (0). The count plot presents the number of customers falling into each purchase category.

### Question 2: Number of Brands Customers Typically Buy (Brand and Quantity)

Next, we explore the number of brands customers typically buy and the quantity of items they purchase. We use two bar plots to display the count of customers for each brand and the count of customers for each quantity of items bought.

### Question 3: Relationship between Current and Previous Purchases (Brand and Last_Inc_Brand)

To investigate the relationship between current and previous purchases made by customers, we group the data by customer ID and calculate the mean of the grouped data. Then, we create a scatter plot that visualizes the current brand purchased on the x-axis and the brand purchased on the previous store visit on the y-axis.

### Question 5: Influence of Promotions on Customer Buying Behavior (Promotion_1 to Promotion_5)

We proceed by analyzing the influence of promotions on customer buying behavior for each promotion column (Promotion_1 to Promotion_5). By grouping the data by customer ID and calculating the mean of the grouped data, we obtain the proportion of customers influenced by promotions for each brand. The resulting bar plot will illustrate the proportion of customers influenced by promotions.

These questions guide us through a comprehensive Exploratory Data Analysis (EDA), providing valuable insights into the purchasing behavior of customers in the chocolate candy bar category. The visualizations offer a clear and concise representation of the data, empowering businesses to make data-driven decisions based on the observed patterns and trends.

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib.axes as axs
import seaborn as sns
sns.set()

pd.options.display.max_columns = None
```

```python
df_ = pd.read_csv('../data/raw/purchase data.csv')
data = df_.copy()
```

## Data Preparation

```python
grouped_data = data.groupby('ID')
```

```python
grouped_data
```

```python

data.head()
```

```python

# Question 1: Overall distribution of purchases (Incidence)
plt.figure(figsize=(6, 4))
sns.countplot(x='Incidence', data=data)
plt.title('Incidence Distribution')
plt.xlabel('Purchase Incidence')
plt.ylabel('Count')
plt.savefig('../reports/figures/incidence_distribution.png')
plt.show()
```

```python

# Question 2: Number of brands customers typically buy (Brand and Quantity)
brand_count = data['Brand'].value_counts()
quantity_count = data['Quantity'].value_counts()

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
sns.barplot(x=brand_count.index, y=brand_count.values)
plt.title('Brand Count')
plt.xlabel('Brand')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
sns.barplot(x=quantity_count.index, y=quantity_count.values)
plt.title('Quantity Count')
plt.xlabel('Quantity')
plt.ylabel('Count')

plt.tight_layout()
plt.savefig('../reports/figures/brand_quantity_count.png')
plt.show()
```

```python


grouped_data = data.groupby('ID').mean()

# Question 3: Relationship between current and previous purchases (Brand and Last_Inc_Brand)
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Brand', y='Last_Inc_Brand', data=grouped_data)
plt.title('Current vs. Previous Brand Purchase')
plt.xlabel('Current Brand')
plt.ylabel('Previous Brand')
plt.savefig('../reports/figures/promotion_influence.png')
plt.show()
```

```python

# Question 4: Variation of prices among different brands (Price_1 to Price_5)
price_cols = ['Price_1', 'Price_2', 'Price_3', 'Price_4', 'Price_5']
plt.figure(figsize=(8, 5))
sns.boxplot(data=grouped_data[price_cols])
plt.title('Price Distribution for Different Brands')
plt.xlabel('Brand')
plt.ylabel('Price')
plt.savefig('../reports/figures/distribution_of_brands.png')
plt.show()
```

```python

# Group the data by ID and calculate the mean of the grouped data
grouped_data = data.groupby('ID').mean()

# Question 5: Influence of promotions on customer buying behavior (Promotion_1 to Promotion_5)
promotion_cols = ['Promotion_1', 'Promotion_2', 'Promotion_3', 'Promotion_4', 'Promotion_5']
promotion_prop = grouped_data[promotion_cols].mean()

plt.figure(figsize=(8, 5))
sns.barplot(x=promotion_prop.index, y=promotion_prop)
plt.title('Influence of Promotions on Purchase')
plt.xlabel('Promotion')
plt.ylabel('Proportion of Customers')
plt.savefig('../reports/figures/influence_of_promotion.png')
plt.show()
```
