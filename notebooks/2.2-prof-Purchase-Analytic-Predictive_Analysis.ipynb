# Purchase Data - Legend

The dataset consists of valuable information regarding the purchases of chocolate candy bars made by 500 individuals in a physical 'FMCG' store over a 2-year period. The data has been collected through loyalty cards used at checkout and has undergone preprocessing, resulting in no missing values. Additionally, to safeguard customer privacy, the dataset's volume has been restricted and anonymized.

## Variable Descriptions

1. **ID**: Numerical - Shows a unique identifier of a customer.
2. **Day**: Numerical - Represents the day when the customer visited the store.
3. **Incidence**: Categorical {0,1} - Indicates the purchase incidence.
    - 0: The customer has not purchased an item from the category of interest.
    - 1: The customer has purchased an item from the category of interest.
4. **Brand**: Categorical {0,1,2,3,4,5} - Shows the brand ID that the customer has purchased.
    - 0: No brand was purchased.
    - 1, 2, 3, 4, 5: Brand ID.
5. **Quantity**: Numerical (Integer) - Represents the number of items bought by the customer from the product category of interest.
6. **Last_Inc_Brand**: Categorical {0,1,2,3,4,5} - Indicates the brand ID that the customer purchased on their previous store visit.
    - 0: No brand was purchased.
    - 1, 2, 3, 4, 5: Brand ID.
7. **Last_Inc_Quantity**: Numerical (Integer) - Represents the number of items bought by the customer from the product category of interest during their previous store visit.
8. **Price_1**: Numerical (Real) - Represents the price of an item from Brand 1 on a particular day.
9. **Price_2**: Numerical (Real) - Represents the price of an item from Brand 2 on a particular day.
10. **Price_3**: Numerical (Real) - Represents the price of an item from Brand 3 on a particular day.
11. **Price_4**: Numerical (Real) - Represents the price of an item from Brand 4 on a particular day.
12. **Price_5**: Numerical (Real) - Represents the price of an item from Brand 5 on a particular day.
13. **Promotion_1**: Categorical {0,1} - Indicates whether Brand 1 was on promotion on a particular day.
    - 0: There is no promotion.
    - 1: There is a promotion.
14. **Promotion_2**: Categorical {0,1} - Indicates whether Brand 2 was on promotion on a particular day.
    - 0: There is no promotion.
    - 1: There is a promotion.
15. **Promotion_3**: Categorical {0,1} - Indicates whether Brand 3 was on promotion on a particular day.
    - 0: There is no promotion.
    - 1: There is a promotion.
16. **Promotion_4**: Categorical {0,1} - Indicates whether Brand 4 was on promotion on a particular day.
    - 0: There is no promotion.
    - 1: There is a promotion.
17. **Promotion_5**: Categorical {0,1} - Indicates whether Brand 5 was on promotion on a particular day.
    - 0: There is no promotion.
    - 1: There is a promotion.
18. **Sex**: Categorical {0,1} - Represents the biological sex (gender) of a customer.
    - 0: Male.
    - 1: Female.
19. **Marital Status**: Categorical {0,1} - Indicates the marital status of a customer.
    - 0: Single.
    - 1: Non-single (divorced / separated / married / widowed).
20. **Age**: Numerical (Integer) - Represents the age of the customer in years, calculated as the current year minus the year of birth of the customer at the time of dataset creation.
    - Min value: 18 (the lowest age observed in the dataset).
    - Max value: 75 (the highest age observed in the dataset).
21. **Education**: Categorical {0,1,2,3} - Represents the level of education of the customer.
    - 0: Other / Unknown.
    - 1: High school.
    - 2: University.
    - 3: Graduate school.
22. **Income**: Numerical (Real) - Represents the self-reported annual income in US dollars of the customer.
    - Min value: 38,247 (the lowest income observed in the dataset).
    - Max value: 309,364 (the highest income observed in the dataset).
23. **Occupation**: Categorical {0,1,2} - Represents the category of occupation of the customer.
    - 0: Unemployed / Unskilled.
    - 1: Skilled employee / Official.
    - 2: Management / Self-employed / Highly qualified employee / Officer.
24. **Settlement Size**: Categorical {0,1,2} - Represents the size of the city that the customer lives in.
    - 0: Small city.
    - 1: Mid-sized city.
    - 2: Big city.
